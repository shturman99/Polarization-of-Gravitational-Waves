# Model Comparison

Stationary-versus-decaying comparisons and coarse scan workflows.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

from compute_H import H_pq, H_pq_decaying, H_pq_decaying_grid

OUT = Path("../outputs")
OUT.mkdir(exist_ok=True)
M, R, k0 = 0.1, 1e4, 1.0
ps = np.logspace(-2, 0.8, 12)
qs = np.logspace(-2, 0.8, 12)
```

## Stationary Grid

Use a coarse grid first. Increase resolution only after the behavior is stable.

```python
H_stationary = np.zeros((len(qs), len(ps)))
for i, q in enumerate(qs):
    for j, p in enumerate(ps):
        H_stationary[i, j] = H_pq(p, q, M=M, R=R, k0=k0, epsabs=5e-4, epsrel=3e-3)

plt.figure(figsize=(6, 4))
plt.pcolormesh(ps, qs, H_stationary, shading="auto")
plt.xscale("log")
plt.yscale("log")
plt.colorbar(label="H stationary")
plt.tight_layout()
plt.show()
```

## Decaying Grid

The packaged helper wraps error handling and is the preferred route for coarse decaying scans.

```python
H_decaying = H_pq_decaying_grid(ps, qs, M=M, R=R, k0=k0, verbose=True)

plt.figure(figsize=(6, 4))
plt.pcolormesh(ps, qs, H_decaying, shading="auto")
plt.xscale("log")
plt.yscale("log")
plt.colorbar(label="H decaying")
plt.tight_layout()
plt.show()
```

## Side-by-Side Comparison

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
pcm = ax[0].pcolormesh(ps, qs, H_stationary, shading="auto")
ax[0].set_xscale("log")
ax[0].set_yscale("log")
ax[0].set_title("stationary")
ax[1].pcolormesh(ps, qs, H_decaying, shading="auto")
ax[1].set_xscale("log")
ax[1].set_yscale("log")
ax[1].set_title("decaying")
for axis in ax:
    axis.set_xlabel("p")
    axis.set_ylabel("q")
fig.colorbar(pcm, ax=ax.ravel().tolist(), label="H")
fig.tight_layout()
plt.show()
```
